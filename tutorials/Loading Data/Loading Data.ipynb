# Visualise

```python
eeg, spec, vote = rLoad[420]
```

```python
eeg = eeg.reset_index(drop=True)
spec = spec.reset_index(drop=True)
```

```python
nrows = 5
ncols = 4
fig, ax = plt.subplots(nrows, ncols, figsize=(20, 20))

for row in range(nrows):
    for col in range(ncols):
        index = row * ncols + col
        ts = eeg.iloc[:, index]
        ax_ = ax[row, col]
        ax_.plot(ts.index.values / 200, ts)
        ax_.set_title(ts.name)
```

```python
ll = spec.iloc[:, :100]
rl = spec.iloc[:, 100:200]
lp = spec.iloc[:, 200:300]
rp = spec.iloc[:, 300:400]
```

```python
ll_freqs = [float(colname[3:]) for colname in ll.columns]
rl_freqs = [float(colname[3:]) for colname in rl.columns]
lp_freqs = [float(colname[3:]) for colname in lp.columns]
rp_freqs = [float(colname[3:]) for colname in rp.columns]
```

```python
spec_data = [
    ("Left Lateral",       ll, ll_freqs, ll.index.values * 2),
    ("Right Lateral",      rl, rl_freqs, rl.index.values * 2),
    ("Left Parasagittal",  lp, lp_freqs, lp.index.values * 2),
    ("Right Parasagittal", rp, rp_freqs, rp.index.values * 2),
]
```

```python
nrows = 2
ncols = 2
fig, ax = plt.subplots(nrows, ncols, figsize=(10, 10))

for row in range(nrows):
    for col in range(ncols):
        index = row * ncols + col
        name, region_spec, freqs, time_index = spec_data[index]
        ax_ = ax[row, col]
        ax_.imshow(region_spec.T, aspect="auto", origin="lower", extent=[time_index[0], time_index[-1], freqs[0], freqs[-1]])
        ax_.set_xlabel("Time in Seconds")
        ax_.set_ylabel("Frequency in Hz")
        ax_.set_title(name)
```
